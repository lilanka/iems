```python
# powerfactor of each load
pf = np.cos(np.arctan(net.load['q_mvar']/net.load['p_mw']))
pf
```

```python
# Use these powerfactors for loads, when assigning P of loads.
pf_loads_forP = np.array([0.98, 0.97, 0.97, 0.97, 0.97, 0.97, 0.97, 0.97, 0.98, 0.97, 0.95, 0.85, 0.85, 0.85, 0.85, 0.85, 0.85, 0.85])
pf_loads_forP
```

```python
# Use these powerfactors for loads, when assigning Q of loads.
pf_loads_forQ = np.sin(np.arccos(pf_loads_forP))
pf_loads_forQ
```
